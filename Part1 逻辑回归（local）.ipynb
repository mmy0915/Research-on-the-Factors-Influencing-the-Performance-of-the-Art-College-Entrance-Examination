```python
def sensitivity_my(tp,fn):
    return tp / (tp + fn)
```

```python
def specificity_my(tn,fp):
    return tn / (tn + fp)
```

```python
def kappa(tp,tn,fp,fn):
    total_accuracy = (tp + tn)/(tp + tn + fp + fn)
    random_accuracy = ((tn + fp) * (tn + fn) + (fn + tp) * (fp + tp))/((tp + tn + fp + fn)  * (tp + tn + fp + fn))
    return (total_accuracy - random_accuracy)/(1 - random_accuracy)
```

```python
def Youden_Index(tn,tp,fn,fp):
    ans = sensitivity_my(tp,fn) - 1.0 + specificity_my(tn,fp)
    return ans
```

```python
def g_mean(tn,tp,fn,fp):
    return math.sqrt(sensitivity_my(tp,fn) * specificity_my(tn,fp))
```

```python
def mcc(tn,tp,fn,fp):
    ans = (tp * tn - fp * fn)/math.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
    return ans
```

```python
def balanced_accuracy(tn,tp,fn,fp):
    ans = 0.5 * sensitivity_my(tp,fn) * specificity_my(tn,fp)
    return ans
```

```python
def precision(tp,fp):
    return tp / (tp + fp)
```

```python
def f_measure(tn,tp,fn,fp):
    up = 2.0 * sensitivity_my(tp,fn) * precision(tp,fp)
    down = sensitivity_my(tp,fn) + precision(tp,fp)
    return up / down
```

```python
def AFM(tn,tp,fn,fp,tn_i,tp_i,fn_i,fp_i):
    f2 = 5.0 * sensitivity_my(tp,fn) * precision(tp,fp) / (4.0 * sensitivity_my(tp,fn) + precision(tp,fp))
    inv_f = 5.0 / 4.0 * sensitivity_my(tp_i,fn_i) * precision(tp_i,fp_i) / (0.5 * 0.5 * sensitivity_my(tp_i,fn_i) + precision(tp_i,fp_i))
    return math.sqrt(f2 * inv_f)
```

```python
def lift(tn,tp,fn,fp):
    up = tp / (tp + fp)
    down = (tp + fn) / (tp + tn + fp + fn)
    return up / down
```

```python
def fpr_my(fp,tn):
    return fp / (fp + tn)
```
